# Working With Matplotlib Module
    
    '''
       
       
       Matplotlib Architecture
          1. Backend Layer(FigureCanvas, Render, Event)
          2. Artist Layer (Artist)
          3. Scripting Layer (Pyplot)
          
          Now About the Each Layer: 
            
            1.Backend Layer --> has 3 built-in abstract interface classes
                                  
                                  1.FigureCancas -- matplotlib.backend_bases.FigureCanvas
                                  
                                    --> Encompresses the area onto which the figure is drawn.
                                  
                                  2.Render       -- matplotlib.backend_bases.Render
                                  
                                    --> Knows how to draw on the FigureCanvas
                                    
                                  3.Event        -- matplotlib.backend_bases.Event
                                  
                                    --> Handles user inputs such keyboard strokes and mouse click.
            
            2. Artsist Layer --> Comprised of one main object-- Artist
                 
                 --> Knows how to use the Renderer to draw on the canvas
                 --> Title, lines,ticklables, images all corresponds to individual Artist instances
                 --> it has 2 types of Artist Object:
                        1.Primitive : Line2D, Rectangle, Circle and Text
                        2.Composit  : Axis, Tick, Axes, Figure
            
            3. Scripting Layer --> Pyplot
    
    '''

```python
### We want the subplots which have 2 row and 2 column
fig, ax = plt.subplots(2,2)  # this will make 2-D subplots means 1 row and 2 columns

# to access 2-D subplots u need to passed the index as tuple type
# like we want the first graph then , choose sheet= 0 and then column = 0
ax[(0, 0)].plot(x)

# we want now sheet = 0, column = 1
ax[(0,1)].scatter(x, y)

#we want now sheet = 1, column = 0
ax[(1, 0)].bar(x, y)

#we want now sheet = 1, column = 1
ax[(1, 1)].hist(z)


# now set the labels

# setting for first graph
ax[(0,0)].set_xlabel('x-axis')
ax[(0,0)].set_ylabel('y-axis')
ax[(0,0)].set_title('Line Graph')

# setting for second graph
ax[(0,1)].set_xlabel('x-axis')
ax[(0,1)].set_ylabel('y-axis')
ax[(0,1)].set_title('Scatter Graph')

# setting for first graph
ax[(1,0)].set_xlabel('x-axis')
ax[(1,0)].set_ylabel('y-axis')
ax[(1,0)].set_title('Bar Graph')

# setting for first graph
ax[(1,1)].set_xlabel('x-axis')
ax[(1,1)].set_ylabel('y-axis')
ax[(1,1)].set_title('Histrogram Graph')

plt.tight_layout()
```

# Working With Tick Marks

    '''
         --> Tick Marks has 2 types :
                                    1. Major --> See on the x and y axis normally have those ticks
                                    2. Minor --> the ticks inside the major ticks
        
        ---> Tick Marks has 2 propeperties :
                                    1. Locator
                                    2. Foramatter
        ---> Has Some Methods along with it,
                      
                      1.set_xticks()
                      2.set_yticks()
                        ---> pass the list of numbers to set the co-ordinates for the ticks
                      3.set_xticklabels()
                      4.set_yticklabels()
                        --> pass the list of strings to set the labels for the ticks

    '''

### Working with the Locator Ticks

```python
### We want the subplots which have 2 row and 2 column
x = np.linspace(1, 100, 5)
y = np.linspace(1, 100, 5)
fig, ax = plt.subplots(1,2)  # this will make 1-D subplots means 1 row and 2 columns
ax[0].plot(x, y)
ax[1].bar(y, x)

#setting the labels

#setting for x axis
ax[0].set_xticks(range(0, 110, 10))
ax[1].set_xticks(range(0, 110, 10))

#setting for y axis
#setting for x axis
ax[0].set_yticks(range(0, 110, 5))
ax[1].set_yticks(range(0, 110, 5))


plt.tight_layout()
```

### Working  with the Formatter Label

```python
### We want the subplots which have 2 row and 2 column
x = np.linspace(1, 100, 5)
y = np.linspace(1, 100, 5)
fig, ax = plt.subplots(1,2)  # this will make 1-D subplots means 1 row and 2 columns
ax[0].plot(x, y)
ax[1].bar(y, x)

#setting the location

#setting for x axis
ax[0].set_xticks(range(0, 110, 10))
ax[1].set_xticks(range(0, 110, 10))

#setting for y axis
ax[0].set_yticks(range(0, 110, 10))
ax[1].set_yticks(range(0, 110, 10))

# setting the foramatter labels

#setting the formatter for axis 
ax[0].set_xticklabels(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'])
ax[1].set_yticklabels(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'])

plt.tight_layout()
```

### Working With Location Class and Formatter Class

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
### We want the subplots which have 2 row and 2 column
x = np.linspace(1, 100, 5)
y = np.linspace(1, 100, 5)
fig, ax = plt.subplots(1,2)  # this will make 1-D subplots means 1 row and 2 columns
ax[0].plot(x, y, color = 'g')
ax[1].bar(y, x, color = 'm')

## Setting the location class Major Ticks On Subplot 1
#setting for x axis major ticks
ax[0].xaxis.set_major_locator(tick.MultipleLocator(base = 20))

#setting y axis major ticks
ax[0].yaxis.set_major_locator(tick.FixedLocator([15, 30, 45, 60, 75, 105]))


## Setting the location class Minor Ticks
#setting for x-axis minor ticks
ax[0].xaxis.set_minor_locator(tick.MultipleLocator(base = 10))

#setting y-axis minor ticks
ax[0].yaxis.set_minor_locator(tick.MultipleLocator(base = 0.5))


## Setting the foramatter labels on Subplot2
#setting the formatter for x-axis 
ax[1].xaxis.set_major_formatter(tick.FormatStrFormatter('%1.2f'))
#setting the formatter for y-axis
ax[1].yaxis.set_major_formatter(tick.FixedFormatter([15, 30, 45, 60, 75, 100]))


plt.tight_layout()
```

# Working With Polar Plotting

### Polar Plotting Using Single  Variable

```python
import matplotlib.pyplot as plt
x = [10, 230, 40, 450, 30, 13, 32, 90, 32, 12, 9, 5, 35, 5, 50, 7]

plt.polar(x)
plt.title('Box Plotting Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

### Polar Plotting Using Multiple Variable

```python
import matplotlib.pyplot as plt
x = [10, 230, 40, 450, 30, 13, 32, 90, 32, 12, 9, 5, 35, 5, 50, 7]
y = [3, 23, 4, 23, 4, 23, 12, 13, 42, 24, 53, 343, 75, 231, 86, 75]


plt.polar(x, y)
plt.title('Box Plotting Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
